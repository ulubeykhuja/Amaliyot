### **Vazifalar**

1. **Maʼlumotlarni yuklash va tozalash**

**Mushroom dataset ustunlari va ularning tavsifi**  

1. **class** – qoʻziqorinning turi: zaharli (`p`) yoki zaharsiz (`e`). *Bu target (maqsad) ustun hisoblanadi.*  
2. **cap-shape** – qoʻziqorin shapkasi shakli:  
   - `b`: bell (qo‘ng‘iroq)  
   - `c`: conical (konussimon)  
   - `x`: convex (yumaloq)  
   - `f`: flat (tekis)  
   - `k`: knobbed (dumaloq tugunli)  
   - `s`: sunken (botiq)  

3. **cap-surface** – qoʻziqorin shapkasi yuzasi qanday koʻrinishda:  
   - `f`: fibrous (tolali)  
   - `g`: grooved (egik chiziqli)  
   - `y`: scaly (taroqli)  
   - `s`: smooth (silliq)  

4. **cap-color** – qoʻziqorin shapkasi rangi:  
   - `n`: brown (jigarrang)  
   - `b`: buff (to‘q sariq-jigarrang)  
   - `c`: cinnamon (dolchin rang)  
   - `g`: gray (kulrang)  
   - `r`: green (yashil)  
   - `p`: pink (pushti)  
   - `u`: purple (binafsha)  
   - `e`: red (qizil)  
   - `w`: white (oq)  
   - `y`: yellow (sariq)  

5. **bruises** – qoʻziqorin bosilganda dogʻ qoldiradimi?  
   - `t`: true (ha)  
   - `f`: false (yo‘q)  

6. **odor** – qoʻziqorin hidi:  
   - `a`: almond (bodom)  
   - `l`: anise (anis)  
   - `c`: creosote (kreozot – o‘ziga xos oʻtkir hid)  
   - `y`: fishy (baliq hidi)  
   - `f`: foul (yomon hid)  
   - `m`: musty (mog‘or hidi)  
   - `n`: none (hidsiz)  
   - `p`: pungent (keskin hid)  
   - `s`: spicy (achchiq hid)  

7. **gill-attachment** – qoʻziqorin plastinkalarining shapkaga bogʻlanish turi:  
   - `a`: attached (biriktirilgan)  
   - `d`: descending (pastga tushgan)  
   - `f`: free (erkin)  
   - `n`: notched (egik)  

8. **gill-spacing** – plastinkalar orasidagi masofa:  
   - `c`: close (yaqin)  
   - `w`: crowded (juda zich)  
   - `d`: distant (uzoq)  

9. **gill-size** – plastinkalarning o‘lchami:  
   - `b`: broad (keng)  
   - `n`: narrow (tor)  

10. **gill-color** – plastinkalarning rangi:  
    - `k`: black (qora)  
    - `n`: brown (jigarrang)  
    - `b`: buff (to‘q sariq-jigarrang)  
    - `h`: chocolate (shokolad rang)  
    - `g`: gray (kulrang)  
    - `r`: green (yashil)  
    - `o`: orange (to‘q sariq)  
    - `p`: pink (pushti)  
    - `u`: purple (binafsha)  
    - `e`: red (qizil)  
    - `w`: white (oq)  
    - `y`: yellow (sariq)  

11. **stalk-shape** – qoʻziqorin poyasi shakli:  
    - `e`: enlarging (yuqoriga qarab kengaygan)  
    - `t`: tapering (yuqoriga qarab toraygan)  

12. **stalk-root** – qoʻziqorin poyasining ildiz qismi:  
    - `b`: bulbous (kurtakli)  
    - `c`: club (to‘qmoqsimon)  
    - `u`: cup (chashka shaklida)  
    - `e`: equal (baravar)  
    - `z`: rhizomorphs (ildiz tolalar bilan bog‘langan)  
    - `r`: rooted (ildizlangan)  
    - `?`: nomaʼlum  

13. **stalk-surface-above-ring** – qoʻziqorin poyasining halqadan yuqoridagi yuzasi:  
    - `f`: fibrous (tolali)  
    - `y`: scaly (taroqli)  
    - `k`: silky (ipaksimon)  
    - `s`: smooth (silliq)  

14. **stalk-surface-below-ring** – qoʻziqorin poyasining halqadan pastdagi yuzasi:  
    - `f`: fibrous (tolali)  
    - `y`: scaly (taroqli)  
    - `k`: silky (ipaksimon)  
    - `s`: smooth (silliq)  

15. **stalk-color-above-ring** – halqadan yuqoridagi poya rangi:  
    - `n`: brown (jigarrang)  
    - `b`: buff (to‘q sariq-jigarrang)  
    - `c`: cinnamon (dolchin rang)  
    - `g`: gray (kulrang)  
    - `o`: orange (to‘q sariq)  
    - `p`: pink (pushti)  
    - `e`: red (qizil)  
    - `w`: white (oq)  
    - `y`: yellow (sariq)  

16. **stalk-color-below-ring** – halqadan pastdagi poya rangi:  
    - `n`: brown (jigarrang)  
    - `b`: buff (to‘q sariq-jigarrang)  
    - `c`: cinnamon (dolchin rang)  
    - `g`: gray (kulrang)  
    - `o`: orange (to‘q sariq)  
    - `p`: pink (pushti)  
    - `e`: red (qizil)  
    - `w`: white (oq)  
    - `y`: yellow (sariq)  

17. **veil-type** – qoʻziqorin pardasi turi (barcha qiymatlar `p`, yaʼni **partial** (qisman) boʻlgani uchun foydasiz ustun).  

18. **veil-color** – parda rangi:  
    - `n`: brown (jigarrang)  
    - `o`: orange (to‘q sariq)  
    - `w`: white (oq)  
    - `y`: yellow (sariq)  

19. **ring-number** – poyada nechta halqa bor:  
    - `n`: none (yo‘q)  
    - `o`: one (bitta)  
    - `t`: two (ikkita)  

20. **ring-type** – halqa turi:  
    - `c`: cobwebby (o‘rgimchak to‘ri kabi)  
    - `e`: evanescent (tez yo‘qoladigan)  
    - `f`: flaring (kengaygan)  
    - `l`: large (katta)  
    - `n`: none (yo‘q)  
    - `p`: pendant (osilgan)  
    - `s`: sheathing (o‘rab turuvchi)  
    - `z`: zone (halqa shaklidagi chiziq)  

21. **spore-print-color** – qoʻziqorin sporalarining rangi:  
    - `k`: black (qora)  
    - `n`: brown (jigarrang)  
    - `b`: buff (to‘q sariq-jigarrang)  
    - `h`: chocolate (shokolad rang)  
    - `r`: green (yashil)  
    - `o`: orange (to‘q sariq)  
    - `u`: purple (binafsha)  
    - `w`: white (oq)  
    - `y`: yellow (sariq)  

22. **population** – qoʻziqorinlarning soni:  
    - `a`: abundant (ko‘p)  
    - `c`: clustered (to‘plangan)  
    - `n`: numerous (juda ko‘p)  
    - `s`: scattered (sochilgan)  
    - `v`: several (bir necha)  
    - `y`: solitary (yakka)  

23. **habitat** – qoʻziqorin yashaydigan muhit:  
    - `g`: grasses (o‘tloq)  
    - `l`: leaves (barglar orasida)  
    - `m`: meadows (yaylovlar)  
    - `p`: paths (yo‘l bo‘ylab)  
    - `u`: urban (shahar atrofida)  
    - `w`: waste (tashlandiq joylar)  
    - `d`: woods (o‘rmon)
