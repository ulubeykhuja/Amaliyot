#**6) APIÊ¼lar bilan ishlash**
1. Yuqorida berilgan kriptovalyuta maÊ¼lumotlaridan `id` ustunini roÊ»yxat koÊ»rinishida saqlang va bu roÊ»yxatga `coin_ids` deb nom bering.
2. https://api.coingecko.com/api/v3/simple/price?ids={coin_ids}&vs_currencies=usd havolasi orqali kriptovalyutalarning `usd` ga nisbatini oling (birinchi 100 ta maÊ¼lumotni olish yetarli).
3. 2-shartda berilgan link orqali olingan maÊ¼lumotlarni dataframe koÊ»rinishida saqlang. Bunda uning indeksi va `usd` ustuni boÊ»lsin. `index` -> kriptovalyutalar nomi, `usd` -> narxi

#**ðŸ’¡Ishora:** linkdagi `{coin_ids}` oÊ»zgaruvchi sifatida yozilgan. Siz ham 1-shartda saqlagan `coin_ids` oÊ»zgaruvchingizni shu qismda yozib keting!


````python
#javobni shu yerda yozing.

# 1. Barcha kriptovalyutalar ro'yxatini olish

coin_ids =  "https://api.coingecko.com/api/v3/coins/list"
response = requests.get(coins_url)
df = pd.DataFrame(response.json())

# 2. Birinchi 100 ta coin id'ni olish
coin_ids = df["id"].head(100).tolist()
coin_ids_str = ",".join(coin_ids)

# 3. Ularning USD narxlarini olish
price_url = f"https://api.coingecko.com/api/v3/simple/price?ids={coin_ids_str}&vs_currencies=usd"
price_data = requests.get(price_url).json()

# 4. DataFrame yaratish
price_df = pd.DataFrame(price_data).T
price_df.columns = ["usd"]

# 5. Natijani koâ€˜rsatish
print(price_df.head())
````

#**7) Veb sahifadan oÊ»qish**
1. Pandas yordamida veb sahifadagi jadvalni yuklab oling.
2. Maâ€™lumotlarni dataframe koâ€˜rinishida chiqaring.
3. Agar sahifada bir nechta jadval boâ€˜lsa, eng muhimini tanlang.
4. Maâ€™lumotni tozalang (boâ€˜sh qatorlarni olib tashlang, kerakli ustunlarni tanlang).

***Eslatma: manba sifatida oÊ»zingiz istagan veb sahifani tanlang. Bu vazifada mustaqil izlanishni oÊ»rganasiz!***

````python
#javobni shu yerda yozing.

from bs4 import BeautifulSoup

#Jadval joylashgan URL manzil
url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0 Safari/537.36"
}

#HTML sahifani yuklab olish
response = requests.get(url, headers=headers)

#HTML kodni BeautifulSoup yordamida oâ€˜qish
soup = BeautifulSoup(response.text, "html.parser")

#Pandas yordamida barcha jadvallarni olish
tables = pd.read_html(str(soup))
print(f"Topilgan jadvallar soni: {len(tables)}")

#Eng muhim jadval â€” birinchisini tanlaymiz
df = tables[2]
print("\nJadvaldan birinchi 5 ta qator:")
print(df.head())

#cleaning
df = df.dropna(how="all")
df = df.rename(columns={
    "Country/Territory": "Country",
    "IMF (2025)[6]": "IMF",
    "World Bank (2022â€“24)[7]": "WB",
    "United Nations (2023)[8]": "UN"
})

df = df[["Country", "IMF", "WB", "UN"]]

# ta_csv
df.to_csv("byGDP.csv", index=False)
````

#**9) MaÊ¼lumotlar omboriga soÊ»rov yuborish**
1. Jadvallar roÊ»yxatini koÊ»ring. Ularning ichida `albums` jadvali boÊ»lsa, shu jadvaldagi barcha maÊ¼lumotlarni (qatorlarni) ekranga chiqaring.
2. Qatorlar soni nechta ekanini aniqlab, uni ham ekranga chiqaring.
3.`albums` jadvalining ustunlari haqidagi maÊ¼lumotlarni ekranga chiqaring.


````python
#javobni shu yerda yozing.

# Ma'lumotlar omboriga ulanish
connection = sqlite3.connect("chinook.db")
cursor = connection.cursor()

# Bazadagi jadvallar ro'yxatini ko'rish
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = [t[0] for t in cursor.fetchall()]

print("Jadvallar:")
print(tables)

# Agar "albums" jadvali mavjud boâ€˜lsa, maâ€™lumotlarini olish
if "albums" in tables:
    print("\n'albums' jadvali topildi!\n")

    # Barcha qatorlarni olish
    df_albums = pd.read_sql_query("SELECT * FROM albums;", connection)

    # Jadvalni koâ€˜rsatish
    print("'albums' jadvalidagi birinchi 5 ta qator:")
    print(df_albums.head())

    # Qatorlar sonini aniqlash
    row_count = len(df_albums)
    print(f"\nQatorlar soni: {row_count}")

    # Ustunlar haqidagi ma'lumotlarni olish
    cursor.execute("PRAGMA table_info(albums);")
    columns_info = cursor.fetchall()

    print("\n'albums' jadvalining ustunlari:")
    for col in columns_info:
        print(f" - {col[1]} ({col[2]})")
else:
    print("'albums' jadvali topilmadi!")
````

````python
#javobni shu yerda yozing.

query = """
SELECT *
FROM invoices
WHERE (BillingCountry = 'Germany' OR BillingCountry = 'France')
  AND Total > 5;
"""

# Soâ€˜rovni bajarish va DataFrame sifatida olish
df_invoices = pd.read_sql_query(query, connection)

# Natijani koâ€˜rsatish
print("Germaniya yoki Fransiyadan boâ€˜lgan va Total > 5 boâ€˜lgan yozuvlar:")
print(df_invoices.head(10))

# Jami natijalar soni
print(f"\n Jami {len(df_invoices)} ta yozuv topildi.")
````

#**19) BETWEEN and LIKE**

* `customers` jadvalidan ismi `A` harfi bilan boshlanadigan foydalanuvchilarni tanlang.

````python
#javobni shu yerda yozing.

# invoices jadvalidan Total summasi 5 dan 15 gacha boâ€˜lgan yozuvlar
query_between = """
SELECT *
FROM invoices
WHERE Total BETWEEN 5 AND 15;
"""
df_between = pd.read_sql_query(query_between, connection)

print("Total summasi 5 dan 15 gacha boâ€˜lgan invoices yozuvlari:")
print(df_between.head(10))  # faqat birinchi 10 qatorni koâ€˜rsatamiz
print(f"\n Jami {len(df_between)} ta yozuv topildi.\n")

# 2customers jadvalidan ismi 'A' harfi bilan boshlanadigan foydalanuvchilar
query_like = """
SELECT *
FROM customers
WHERE FirstName LIKE 'A%';
"""
df_like = pd.read_sql_query(query_like, connection)

print("Ismi 'A' harfi bilan boshlanadigan foydalanuvchilar:")
print(df_like[["CustomerId", "FirstName", "LastName", "Country"]])  # faqat kerakli ustunlar

# Ulanishni yopish
connection.close()
````
