#1-topshiriq


1.    Quyidagi link orqali maʼlumotlarni Pandas dataframeʼiga yuklab oling.

2.   Yuklab olingan maʼlumotlarni `aqi.csv` fayliga saqlang.

3. Saqlab olingan `aqi.csv` faylining oxirgi 5 ta qatorini oʻqing.

Eslatma: fayl nomini `aqi.csv` deb saqlaganimizning sababi AQI – Air Quality Index, yaʼni havo sifati indeksi deganidir. Bizning maʼlumotlarimizda davlatlarning havo sifati haqidagi maʼlumotlar saqlangan!

```python
import pandas as pd
url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/data_date.csv"
#javobni shu yerda yozing.

df = pd.read_csv(url)

#Yuklab olingan maʼlumotlarni aqi.csv fayliga saqlang.
df.to_csv("aqi.csv", index=False)
#Saqlab olingan aqi.csv faylining oxirgi 5 ta qatorini oʻqing.
aqi_df = pd.read_csv("aqi.csv")
print(aqi_df.tail())
```

#2-topshiriq
1. Saqlab olingan `aqi.csv` faylidagi maʼlumotlarning faqat 2024-yilga tegishlisini excel faylga `aqi_2024.xlsx` koʻrinishida saqlab oling.
2. Saqlab olingan `aqi_2024.xlsx` faylidan Oʻzbekistonga tegishli maʼlumotni (qatorlarni) oʻqing.

```python
#javobni shu yerda yozing.

#faqat 2024-yilga tegishlisini excel faylga aqi_2024.xlsx koʻrinishida saqlab oling.
aqi_df['Date'] = pd.to_datetime(aqi_df['Date'])
aqi_2024 = aqi_df[aqi_df['Date'].dt.year == 2024]
aqi_2024.to_excel("aqi_2024.xlsx", index=False)

#Saqlab olingan aqi_2024.xlsx faylidan Oʻzbekistonga tegishli maʼlumotni (qatorlarni) oʻqing.
aqi_2024_df = pd.read_excel("aqi_2024.xlsx")
uzbekistan_data = aqi_2024_df[aqi_2024_df['Country'] == 'Uzbekistan']

print(uzbekistan_data)
```

#3-topshiriq
1. `aqi.csv` faylidagi maʼlumotlarni oʻrganib chiqsangiz, bu faylda 2022–2025-yillar oraligʻidagi havo sifati haqidagi maʼlumotlarni koʻrishingiz mumkin. Hozir sizga vazifa:
  * `aqi_2022.csv`, `aqi_2023.csv`, `aqi_2024.csv`, `aqi_2025.csv`  fayllariga `aqi.csv` faylidagi maʼlumotlarning mos davrlarini saqlang. Misol uchun, `aqi_2022` fayli 2022-yildagi havo sifati haqidagi maʼlumotlarni oʻzida saqlaydi.
  

```python
#javobni shu yerda yozing.

# date formatiga o'tkazish
aqi_df['Date'] = pd.to_datetime(aqi_df['Date'])

# filtrlash
aqi_2022 = aqi_df[aqi_df['Date'].dt.year == 2022]
aqi_2023 = aqi_df[aqi_df['Date'].dt.year == 2023]
aqi_2024 = aqi_df[aqi_df['Date'].dt.year == 2024]
aqi_2025 = aqi_df[aqi_df['Date'].dt.year == 2025]

# alohida faylda saqlash
aqi_2022.to_csv("aqi_2022.csv", index=False)
aqi_2023.to_csv("aqi_2023.csv", index=False)
aqi_2024.to_csv("aqi_2024.csv", index=False)
aqi_2025.to_csv("aqi_2025.csv", index=False)

#tekshirish
print("2022-yil ma'lumotlari:")
print(aqi_2022.head(3))
print("\n2023-yil ma'lumotlari:")
print(aqi_2023.head(3))
print("\n2024-yil ma'lumotlari:")
print(aqi_2024.head(3))
print("\n2025-yil ma'lumotlari:")
print(aqi_2025.head(3))
```

#4-topshiriq
1. Saqlab olingan `aqi_2022.csv`, `aqi_2023.csv`, `aqi_2024.csv`, `aqi_2025.csv`  fayllarini HDF5 formatida saqlang. Fayl nomini `aqi.h5` deb nomlang. Kalit (key) sifatida fayllarning yillarini bering. Misol uchun, key=2025.

```python
#javobni shu yerda yozing.

#Har bir CSV faylni o‘qiymiz
aqi_2022 = pd.read_csv("aqi_2022.csv")
aqi_2023 = pd.read_csv("aqi_2023.csv")
aqi_2024 = pd.read_csv("aqi_2024.csv")
aqi_2025 = pd.read_csv("aqi_2025.csv")

#HDF5 faylni yaratamiz
with pd.HDFStore("aqi.h5", mode="w") as hdf:
    hdf.put("2022", aqi_2022)
    hdf.put("2023", aqi_2023)
    hdf.put("2024", aqi_2024)
    hdf.put("2025", aqi_2025)

#mavjud kalitlarni ko‘rib chiqamiz
with pd.HDFStore("aqi.h5", mode="r") as hdf:
    print("HDF5 fayldagi mavjud kalitlar:")
    print(hdf.keys())
```

#5-topshiriq
1. `aqi.h5` faylidan maʼlumotlarni oʻqing va 2025-yilga tegishli maʼlumotlarni yangi dataframeʼga saqlang.
2. `aqi.h5` faylida 2025-yilning qaysi oylari uchun maʼlumotlar bor? Shu oylar roʻyxatini ekranga chiqaring!

```python
#javobni shu yerda yozing.

#aqi.h5 faylidan maʼlumotlarni oʻqing va 2025-yilga tegishli maʼlumotlarni yangi dataframeʼga saqlang.
aqi_2025 = pd.read_hdf("aqi.h5", key="2025")

# datetime formatiga o‘tkazamiz
aqi_2025['Date'] = pd.to_datetime(aqi_2025['Date'])

#aqi.h5 faylida 2025-yilning qaysi oylari uchun maʼlumotlar bor?
oylar = sorted(aqi_2025['Date'].dt.month.unique())

#Oy raqamlarini nomlarga o‘giramiz
oy_nomi = {
    1: "Yanvar", 2: "Fevral", 3: "Mart", 4: "Aprel", 5: "May", 6: "Iyun",
    7: "Iyul", 8: "Avgust", 9: "Sentabr", 10: "Oktabr", 11: "Noyabr", 12: "Dekabr"
}
mavjud_oylar = [oy_nomi[o] for o in oylar]

# Natijani chiqarish
print("2025-yil uchun mavjud oylar:")
for oy in mavjud_oylar:
    print("-", oy)
```
