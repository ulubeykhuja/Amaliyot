```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.1, random_state=42)
```

```python
train_set
```

```python
from sklearn import linear_model
LR_model = linear_model.LinearRegression() # Modelni yukladik

x_train = np.asanyarray(train_set[['TV']])  # Model array bilan ishlagani
y_train = np.asanyarray(train_set[['Sales']])# uchun np b-n array qildik

LR_model.fit(x_train, y_train) # Modelni tarbiyalash
```

```python
# teta1 va teta0 koef.larni ajratish
theta1 = LR_model.coef_[0][0]
theta0 = LR_model.intercept_[0]
print(f"{theta1=}")
print(f"{theta0=}")
```
