```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import chi2
from sklearn.ensemble import RandomForestClassifier

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report



url="https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/mushrooms.csv"
df = pd.read_csv(url)
df.head()
```

```python
# 1.2. Kategorik ustunlarni tekshiring, agar kerak boʻlsa, one-hot encoding yoki label encoding qiling.
df.dtypes
```

```python
df['class'].value_counts()
```

```python
labelencoder = LabelEncoder()
df['class'] = labelencoder.fit_transform(df['class'])

df_encoded = pd.get_dummies(df, columns=df.columns[1:])

display(df_encoded.head())
```

```python
# Target (class) кодлаш
le = LabelEncoder()
df['class'] = le.fit_transform(df['class'])  # e->0, p->1

# One-hot encoding barcha qolgan ustunlar uchun
X = df.drop('class', axis=1)
y = df['class']

# ColumnTransformer orqali
ct = ColumnTransformer(
    transformers=[("onehot", OneHotEncoder(drop="first"), X.columns)],
    remainder='drop'
)

X_encoded = ct.fit_transform(X)

print("X_encoded shape:", X_encoded.shape)
print("y shape:", y.shape)
```

```python
# Null qiymatlar mavjudligini tekshiring va kerak bo‘lsa, tozalang.
if hasattr(X_encoded, "toarray"):
    X_check = X_encoded.toarray()
else:
    X_check = X_encoded

print("NaN in X:", np.isnan(X_check).sum())
print("NaN in y:", np.isnan(y).sum())
```

 2. **Maʼlumotlarni tahlil qilish**

1. Ustunlarning tarqalishini vizualizatsiya qiling.
2. Maqsad ustunining balansini tekshiring (nechta zaharli va zaharsiz qo‘ziqorinlar bor).
3. Har bir xususiyatning maqsad ustuniga ta’sirini tahlil qiling.

```python
# 2.1. Ustunlarning tarqalishini vizualizatsiya qiling.

for col in df.columns[:6]:   # misol uchun birinchi 6 ustun
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col, hue="class")
    plt.title(f"{col} tarqalishi (class bo‘yicha)")
    plt.show()
```

```python
# 2.2. Maqsad ustunining balansini tekshiring (nechta zaharli va zaharsiz qo‘ziqorinlar bor).

df['class'].value_counts().plot(kind='bar', color=['green','red'])
plt.title("Maqsad ustuni (class) taqsimoti")
plt.xticks(rotation=0)
plt.xlabel("Class (0=edible, 1=poisonous)")
plt.ylabel("Soni")
plt.show()

print(df['class'].value_counts())
```

```python
# 2.3. Har bir xususiyatning maqsad ustuniga ta’sirini tahlil qiling.
for col in df_encoded.columns:
    if df_encoded[col].dtype == 'object':
        df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col])

X_chi2 = df_encoded.drop("class", axis=1)
y_chi2 = df_encoded["class"]

chi2_scores, p_values = chi2(X_chi2, y_chi2)

chi2_result = pd.DataFrame({
    "Feature": X_chi2.columns,
    "Chi2_score": chi2_scores,
    "p_value": p_values
}).sort_values(by="Chi2_score", ascending=False)

print(chi2_result.head(10))
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_encoded, y)

importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

# Top 15 ta xususiyatni chizish
plt.figure(figsize=(10,6))
plt.bar(range(15), importances[indices[:15]])
plt.xticks(range(15), indices[:15], rotation=90)
plt.title("Eng muhim 15 ta xususiyat (Random Forest)")
plt.show()
```

3. **Modellarni qurish va taqqoslash**
    - Ma’lumotlarni **training** va **test** to‘plamlariga ajrating (80/20 yoki 70/30).
    - Quyidagi klassifikatsiya algoritmlarini qo‘llang va natijalarni solishtiring:
        - **K-Nearest Neighbors (KNN)**
        - **Decision Tree**
        - **Random Forest**
        - **Logistic Regression**
    - Har bir model uchun **Confusion Matrix, Precision, Recall, F1-Score** hisoblang.

```python
# Ma’lumotlarni training va test to‘plamlariga ajrating (80/20 yoki 70/30).
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y,
    test_size=0.2,     # 20% test, 80% train
    random_state=42,   # natijani qayta tiklash uchun
    stratify=y         # target bo‘yicha balansni saqlash
)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

# Quyidagi klassifikatsiya algoritmlarini qo‘llang va natijalarni solishtiring:
- **K-Nearest Neighbors (KNN)**
- **Decision Tree**
- **Random Forest**
- **Logistic Regression**

```python
models = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    results[name] = acc

    print(f"=== {name} ===")
    print("Accuracy:", acc)
    print(classification_report(y_test, y_pred))
    print("\n")
```

```python
# Har bir model uchun Confusion Matrix, Precision, Recall, F1-Score hisoblang.
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt

# Natijalarni saqlash uchun
metrics = []

for name, model in models.items():
    y_pred = model.predict(X_test)

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Edible (0)", "Poisonous (1)"],
                yticklabels=["Edible (0)", "Poisonous (1)"])
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # Metrics
    precision = precision_score(y_test, y_pred)
    recall    = recall_score(y_test, y_pred)
    f1        = f1_score(y_test, y_pred)

    metrics.append({
        "Model": name,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1
    })

# DataFrame ko‘rinishida natijalarni chiqarish
results_df = pd.DataFrame(metrics)
print(results_df)
```
