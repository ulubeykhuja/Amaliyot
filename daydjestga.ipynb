```python
# prompt: как изменит размер шрифта текста кусков

import matplotlib.pyplot as plt

# Данные для круговой диаграммы: распределение потерь энергии по источникам
labels = ['Производство \n энергии', 'Передача и \n распределение', 'Потери на стороне \n потребителя']
sizes = [47, 5, 48]  # в процентах
colors = ['#ff9999', '#66b3ff', '#99ff99']

# Создание круговой диаграммы с увеличенным и жирным заголовком
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 14}) # Изменение размера шрифта текста кусков
plt.title('Распределение потерь \n энергии в мировой системе', fontsize=28, fontweight='bold', pad=1)

plt.axis('equal')  # Круг
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные для диаграммы: распределение инвестиций в чистую энергетику
categories = ['Вся чистая энергетика', 'Энергоэффективность (всего)', 'ЭЭ в LICs и MICs']
values = [100, 20, 8]  # в процентах
colors = ['#1f77b4', '#ff7f0e', '#d62728']

# Построение столбчатой диаграммы
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color=colors)

# Добавим подписи значений
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5), textcoords='offset points',
                 ha='center', va='bottom')

# Настройки графика
plt.title('Доля инвестиций в энергоэффективность в рамках чистой энергетики', fontsize=16, fontweight='bold')
plt.ylabel('Доля, %')
plt.ylim(0, 120)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные для диаграммы: потребность и эффект от льготного финансирования
categories = ['Необходимое льготное финансирование', 'Потенциальная экономия до 2050 г.']
values = [6, 11.6]  # в млрд долларов
colors = ['#ff7f0e', '#2ca02c']

# Создание столбчатой диаграммы
plt.figure(figsize=(8, 6))
bars = plt.bar(categories, values, color=colors)

# Добавление подписей над столбцами
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height} млрд $',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 5),
                 textcoords="offset points",
                 ha='center', va='bottom', fontsize=14) # Изменение размера шрифта здесь

# Настройки
plt.title('Льготное финансирование и ожидаемая \n экономия от энергоэффективности', fontsize=14, fontweight='bold')
plt.ylabel('Млрд долларов США')
plt.ylim(0, 14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
